```python
import numpy as np
import pandas as pd
import scipy as sp
import matplotlib.pyplot as plt
from numpy.lib.scimath import sqrt
```

```python
# funkcje parametrow pole-2-ground

alpha_1 = lambda C:1/(2*C)
alpha_2 = lambda x,Lx:2/(x*Lx)
alpha_3 = lambda Rf,Rx,x,Lx: alpha_2(x,Lx)*(Rf + x*Rx/2)

s_1 =lambda Rf,Rx,x,Lx: -alpha_3(Rf,Rx,x,Lx)/2 +sqrt((alpha_3(Rf,Rx,x,Lx)/2)**2-alpha_1(C)*alpha_2(x,Lx))
s_2 =lambda Rf,Rx,x,Lx: -alpha_3(Rf,Rx,x,Lx)/2 -sqrt((alpha_3(Rf,Rx,x,Lx)/2)**2-alpha_1(C)*alpha_2(x,Lx))
A = lambda s,x,Lx,V0,I0,Rf,Rx : (alpha_2(x,Lx)*V0+s*I0)/(s_2(Rf,Rx,x,Lx)-s_1(Rf,Rx,x,Lx))
B = lambda s,x,Lx,V0,I0,Rf,Rx : ((s+alpha_3(Rf,Rx,x,Lx) )*V0-alpha_1(C)*I0)/(s_2(Rf,Rx,x,Lx)-s_1(Rf,Rx,x,Lx))
```

```python
#p2g case 3
x=0.5
Rx=0.1
Lx=0.3
C=3
Rf=0.7
V0=0.5
I0=-0.063


print(alpha_1(C))
print(alpha_2(x,Lx))
print(alpha_3(Rf,Rx,x,Lx))
print(s_1(Rf,Rx,x,Lx))
print(s_2(Rf,Rx,x,Lx))
```

```python
s1=s_1(Rf,Rx,x,Lx)
s2=s_2(Rf,Rx,x,Lx)

t=np.linspace(0,5,1000)
i_cable = A(s2,x,Lx,V0,I0,Rf,Rx)*np.exp(s2*t)-A(s1,x,Lx,V0,I0,Rf,Rx)*np.exp(s1*t)
v_c = B(s2,x,Lx,V0,I0,Rf,Rx)*np.exp(s2*t)-B(s1,x,Lx,V0,I0,Rf,Rx)*np.exp(s1*t)

plt.plot(t,i_cable)
plt.plot(t,v_c)
```

```python
# case 1 pole-2-pole
V0=0.5
I0=-0.063

Rx=0.01

Lx=0.3
C=3

x=0.5
Rf=0.4


params_means = np.array([Rx,Lx,C,x,Rf])
params_stds = 0.1 * params_means

randomized = np.random.normal(loc=params_means,scale=params_stds,size=(100,5))
```

```python
#Funkcje parametrów pole-2-pole
alpha_1 = lambda C:1/(C)
alpha_2 = lambda x,Lx:1/(x*Lx)
alpha_3 = lambda Rf,Rx,x,Lx: alpha_2(x,Lx)*(Rf + x*Rx)
t=np.linspace(0,5,100)
currents=[]
voltages=[]
for row in randomized:
    
    Rx=row[0]
    Lx=row[1]
    C=row[2]
    x=row[3]
    Rf=row[4]
    s1=s_1(Rf,Rx,x,Lx)
    s2=s_2(Rf,Rx,x,Lx)
    i_noise = np.random.normal(scale=0.01*np.abs(I0),size=100)
    v_noise = np.random.normal(scale=0.01*V0,size=100)
    i_cable = np.real(A(s2,x,Lx,V0,I0,Rf,Rx)*np.exp(s2*t)-A(s1,x,Lx,V0,I0,Rf,Rx)*np.exp(s1*t))+i_noise
    v_c = np.real(B(s2,x,Lx,V0,I0,Rf,Rx)*np.exp(s2*t)-B(s1,x,Lx,V0,I0,Rf,Rx)*np.exp(s1*t))+v_noise
    currents.append(i_cable)
    voltages.append(v_c)


```

```python
fig,ax = plt.subplots(1,1)
for c,v in zip(currents,voltages):
    ax.plot(t,c,c='b',alpha=0.3)
    ax.plot(t,v,c='r',alpha=0.3)
plt.show()
```
